Getting started with neural networks.

Major components of neural network:
1. Layers
2. Input data and  corresponding targets
3. Loss function
4. Optimizer

```
from keras import layers
layer = layers.Dense(32, input_shape = (784,)) # a dense layer with 32 output units
```

We are creating a layer that will only accept as input 2D tensors where the first dimension is 784 (axis 0, the batch dimension, is unspecified, and thus any value would be accepted). This layer will return a tensor where the first dimension has been transformed to be 32. Hence this layer can only be connected to downstream layer that expects 32-dimensional vectors as its input. 

In keras dimensionality compatibility is not much topic that should be worried of because it automatically handles. For example, shown below:

```
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(32, input_shape = (784,)))
model.add(layers.Dense(32))
```
